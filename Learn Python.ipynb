```python
# b Dictionaries
#These are mutable data structures that contain a collection of keys and, associated with them, values.

print('b) Dictionaries \n')

# Create an empty dictionary using curly brackets
print('empty dictionary \n')
d1 = {}
print(d1)
# simple dictionary
print('a simple dictionary')

d2={'name':'John','age':27,'Hometown':'boston'}
print(d2)
print('the name is: ', d2['name'],'\n')
print ('method 1 all details \n')
for x in d2:
    print(x, ':', d2[x],'\n')
    
print ('method 2 all details \n') 

for key,value in d2.items():
    print(key,':',value,'\n')
# Create an empty dictionary using the dict() constructor
print('Create an empty dictionary using the dict() constructor \n')
d4 = dict()
print(d4)

# Create a two-element dictionary using the dict() constructor
print('Create a  dictionary using the dict() constructor with assigning variables \n')
d5=dict(name1='Nairobi',rank1=1,name2='Mombasa',rank2=2,name3='Eldoret',rank3=3)
print(d5,'\n')
x = dict(name = "John", age = 36, country = "Norway")
print(x,'\n')

# Create a two-element dictionary using the dict() constructor
print('create a two element dict using dict() and in  a list of lists \n')
d6 = dict([["name", 'John'], ["age", 21]])  # Note that we created the dictionary from a list of list

print(type(d6))
print(d6,'\n')


# Create a two-element dictionary using curly brackets
# Note that the  dictionary has a more complex structure as its values are dictionaries themselves!
print('Nested dictionaries')
dn1= {"John": {"Age": 27, "Hometown": "Boston"}, "Rebecca": {"Age": 31, "Hometown": "Chicago"}}
print(dn1)
print('John data is',dn1['John'],'\n')
print('every item in d3 using for loop \n')
for key ,details in dn1.items():
    print(key,'\n')
    for key in details:
        print(key,':',details[key])

people = {1: {'name': 'John', 'age': '27', 'sex': 'Male'},
          2: {'name': 'Marie', 'age': '22', 'sex': 'Female'},
          3: {'name': 'Luna', 'age': '24', 'sex': 'Female', 'married': 'No'}}

print(people,'\n')
print('access individual items')
print(people[1]['name'])
print(people[1]['age'])
print(people[1]['sex'],'\n')

print('Add another dictionary to the nested dict \n')
people[4] = {'name': 'Peter', 'age': '29', 'sex': 'Male', 'married': 'Yes'}
print(people,'\n')
print(people[4],'\n')
print('Iterate on a nested dictionary')

for key ,details in people.items():
    print(key,'\n')
    for key in details:
        print(key,':',details[key])
    
print()
print('delete elements in a dictionary \n')

del people[3]['married']
del people[4]['married']

print(people[3])
print(people[4])
print()
print('delete a dictionary in a nested dictionary \n')

del people[3], people[4]
print(people)

```

```python
# c Sets

print('sets \n')
# Create a set using curly brackets
print('# Create a set using curly brackets \n')

s1 ={1,3,4,7,9}

print(s1)
# Create a set using the set() constructor
s2 = set([8, 10, 19, 4])

print(s2)

print('Sets union()/merging sets \n')

serials={'maize','beans','millet','sorghum'}
fruits={'apple','banana','mangoes','paw paw'}

print('method1 using union() \n')
ser_fruits=serials.union(fruits)

print(ser_fruits,'\n')

print('method2 # Create a union of two sets using the | operator \n')

print(fruits|serials)
print()
print('create sets with similar elements, note only one will appear in a union \n')
fruits1={'apple','banana','mangoes','paw paw'}
fruits2={'apple','orange','mangoes','lemon'}

print('\n',fruits1,'\n',fruits2,'\n')
unionf1andf2=fruits1|fruits2
print(unionf1andf2)

print('Find which elements appear on both sets \n')
print('method1 using the intersection() constructor \n')
interf1f2=fruits1.intersection(fruits2)

print(interf1f2,'\n')

print('method2 using the ampersand (&) operator \n')
print(fruits1&fruits2)

print()
print('return the difference in two sets (which elements are not in the other set) \n')
print('method 1 use difference() constructor \n')

print(fruits1.difference(fruits2))
print()
print('method2 use (-) operator \n')
print(fruits2-fruits1)

```
